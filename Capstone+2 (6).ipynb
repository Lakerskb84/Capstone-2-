```python
dummy=pd.get_dummies(m3[['Gender','Race/Ethnicity']])
m4=pd.concat([m3, dummy], axis=1)
m4
```

```python
m4.rename(columns={'Race/Ethnicity_African American, Not Hispanic':'African American','Race/Ethnicity_All Students':'All Students','Race/Ethnicity_American Indian or Alaska Native, Not Hispanic':'American Indian or Alaska Native','Race/Ethnicity_Asian, Not Hispanic':'Asian','Race/Ethnicity_Filipino, Not Hispanic':'Filipino','Race/Ethnicity_Hispanic or Latino of Any Race':'Hispanic or Latino','Race/Ethnicity_Not Reported':'Not Reported','Race/Ethnicity_Pacific Islander, Not Hispanic':'Pacific Islander','Race/Ethnicity_Two or More Races, Not Hispanic':'Two or More Races','Race/Ethnicity_White, Not Hispanic':'White'}, inplace=True)
m4.info()
```

```python
m5=m4.drop('Gender', axis=1)
m6=m5.drop('Race/Ethnicity', axis=1)
m6.info()
```

```python
m6.shape
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.externals import joblib
```

```python
X=m6.drop(['Year','Cohort Graduation Rate'], axis=1)
y=np.asarray(m6['Cohort Graduation Rate'], dtype='|S6')
#Training and Testing Data

X_train, X_test, y_train, y_test=train_test_split(X, y, train_size=0.20, random_state=42)

#Constructing Pipelines

pipe_lr=Pipeline([('scale',StandardScaler()), ('clf',LogisticRegression(random_state=42))])

pipe_lr_pca=Pipeline([('scale',StandardScaler()),('pca',PCA(n_components=15)),('clf',LogisticRegression(random_state=42))])

pipe_svm=Pipeline([('scale',StandardScaler()),('clf',svm.SVC(random_state=42))])

pipe_svm_pca=Pipeline([('scale',StandardScaler()),('pca',PCA(n_components=15)),('clf',svm.SVC(random_state=42))])

pipe_tree=Pipeline([('sacle',StandardScaler()),('clf',RandomForestClassifier(random_state=42))])

pipe_tree_pca=Pipeline([('scale',StandardScaler()),('pca',PCA(n_components=15)), ('clf',RandomForestClassifier(random_state=42))])

#Parameters for GridSearchCV

param_range=[1,2,3,4,5,]
param_range_fl=[1.0, 0.5, 0.1]

grid_params_lr= [{'clf_penalty':['l2'], 'clf_C':param_range_fl,
                  'clf_solver':['newton-cg'],
                  'clf_multiclass':['multinomial']}]

grid_params_tree=[{'clf_criterion':['gini','entropy'],
                  'clf_min_samples_leaf': param_range,
                  'clf_min_samples_split':param_range[1:]}]

grid_params_svm=[{'clf_kernel':['linear','rbf'],
                 'clf_C': param_range}]

#Constructing Grid Searches

jobs=-1

gs_lr= GridSearchCV(estimator=pipe_lr,
                  param_grid= grid_params_lr,
                  scoring='accuracy',
                  cv=3)

gs_tree= GridSearchCV(estimator=pipe_tree,
                      param_grid=grid_params_tree,
                      scoring='accuracy',
                      cv=3,
                      n_jobs=jobs)

gs_svm= GridSearchCV(estimator=pipe_svm,
                     param_grid= grid_params_svm,
                     scoring='accuracy',
                     cv=3,
                     n_jobs=jobs)



#List of Pipelines

grids= [gs_lr, gs_tree, gs_svm]


#Dictionary of piplelines 

grid_dict= {0: 'LogisticRegression', 1: 'Random Forest', 3: 'svm'} 

# Fit Grid Search Objects

print('Performing model optimizations...')

best_acc=0.0
best_clf=0
best_pipe=''
    
for idx, gs in enumerate(grids):
        print('\nestimator: %s'%grid_dict[idx])
        #fit
        gs.fit(X_train, y_train)
        #get best parameters
        print('Best params:%s'% gs.best_params_)
        #best Training data accuracy
        print('Best Training Data Accuracy: %.3f' % gs.best_score_)
        #Prediction on test set
        y_pred= gs.predict(X_test)
        #Test Accuracy with the best model and best params
        print('Test set accuracy score with the best params: %.3f' % accuracy_score(y_test, y_pred))
        # Highest test accuracy model
        if accuracy_score(y_test, y_pred) > best_acc:
                best_acc= accuracy_score(y_test, y_pred)
                best_gs= gs
                best_clf= idx
print('\nclassifier with best test accuracy: %s'%grid_dict[best_clf])
```

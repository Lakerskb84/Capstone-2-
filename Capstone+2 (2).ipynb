```python
grace='gender_race.csv'
g=pd.read_csv(grace)
print(g)
```

```python
g.info()
```

```python
g1=g[['School Year Ending','Gender','Race/Ethnicity','Cohort Students','Cohort Graduates','Cohort Graduation Rate']]
print(g1)
```

```python
g1['School Year Ending']=pd.to_datetime(g1['School Year Ending'])
g1['School Year Ending'].dt.year
```

```python
g2=g1.groupby(['School Year Ending','Gender'])['Race/Ethnicity','Cohort Students','Cohort Graduates','Cohort Graduation Rate'].mean()
print(g2)
```

```python
g2=g2.reset_index()
print(g2)
```
